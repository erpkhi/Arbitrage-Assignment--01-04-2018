```python
import numpy as np
import pandas as pd
import requests as rq
import json
import matplotlib.pyplot as plt
url='https://api.cryptonator.com/api/full/btc-usd'
data=rq.get(url).json()["ticker"]["markets"]
df2 = pd.DataFrame(data)
maxmarket=df2[df2['price']==df2['price'].max()]['market']
minmarket=df2[df2['price']==df2['price'].min()]['market']

print(str(maxmarket) + ' Has maximum price'+'\n'+ str(minmarket) + 'has minimum value ')
mktlist=df2['market']
pricelist=df2['price']
volume=df2['volume']
max=np.max(df2['price'])
min=np.min(df2['price'])
arbitrage=float(max)-float(min)
print(str(arbitrage) + ' Maximum Posible Profitablty')
plt.scatter(pricelist, volume)
for i, txt in enumerate(mktlist):
    plt.annotate(txt, (pricelist[i],volume[i]))
plt.show()
```
